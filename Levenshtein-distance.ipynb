```python
#import libraries and dependences
import pandas as pd
import distance as dist
```

```python
#upload dataset
profiles =  pd.read_csv('jobseekers.csv')
profiles.head(5)
```

```python
profiles['key_Skills'][0]
```

```python
#let's create a random job post
job_post = {
    'Title':'Technical Analyst - Data Science',
    'Experiencie': "2 - 3 years",
    'key_skills': 
    {       
     'Data Science',
     'Data Analytics',
     'Python'
    },
}
```

```python
# Job post vars
title_job_post = list(job_post['Title'])
experience_job_post = job_post['Experiencie']
skills_job_post = list(job_post['key_skills'])
```

```python
# since 'key_skills' column has joined-words by row, 
# we set a function to split by comparing with a predefined word list 
# (bank_skills below)

bank_skills = ['Data Science', 'Machine Learning','Data Analytics',
              'Artificial Intelligence', 'R', 'Python', 'SQL', 'SAS',
              'Tableau', 'Logistic regression', 'Computer science', 
              'Operations research', 'Analytical Consulting',
              'Business intelligence', 'Azure cloud', 'C++', 
              'Statistical Analysis', 'Anomaly Detection', 'Data Visualization',
              'Statistics', 'Data mining', 'CRM', 'Version control', 'R Shiny',
              'Financial services', 'Data extraction', 'Data Warehousing', 
              'Data quality' , 'Big Data', 'Problem solving', 'Excel' ,
              'Predictive Modeling' , 'Time Series', 'Risk management', 
              'Data science', 'Machine learning', 'SCALA', 'Hive', 'NLP',
              'Open source', 'MATLAB', 'Deep Learning', 'Hadoop', 
              'Data Management', 'Macros']
             # This list to extend yet

# set a function
def split_list_by_words(data_list, word_list):
  words_to_check = []
  for w in word_list:
    if w in data_list:
      words_to_check.append(w)
  return words_to_check
```

```python
#we apply 'split_by_words' function in candidate's skills
all_candidate_skills = profiles['key_Skills']
user_skills = []
user_skills_str = []
for c in range(len(all_candidate_skills)):
  user_skills.append(split_list_by_words(all_candidate_skills[c], 
                                         word_list=bank_skills))
  #for aesthetic purposes we display our results as string 
  user_skills_str.append(','.join(user_skills[c]))
user_skills_str
```

```python
# We've created new column called 'Skills Separate' and was added to 'profiles'
profiles['Skills Separate'] = user_skills_str
profiles['Skills Separate'][0].split(',')
```

```python
# Set funcition which implements Lenvenshtein distance in order to indentify words matches
# Skills coincidences between job post skills and seeker skills
def check_skills(job_skills, candidate_skills):
  """ This function compares jobseekers skills 
  with job offer's. Returns list with Similarity level calculated by
  Levenshtein Distance.
  Example: 
  job_skills = `['Data Science','Data Analytics','Python']`
  candidate_skills = `['Data Science','Machine Learning',
                      'R','Python','SQL','SAS','Tableau',
                      'Statistics','R Shiny']`
  """
  match_skills = []
  for s in range(len(job_skills)):
    match_skills.append(sorted(dist.ilevenshtein(job_skills[s], candidate_skills)))
    words_coincidences = []
    for i in range(len(match_skills)):
      for j in range(len(match_skills[i])):
        for k in range(len(match_skills[i][j])):
          if (match_skills[i][j][k] == 0 or match_skills[i][j][k] == 5 
              or match_skills[i][j][k] == 8 or match_skills[i][j][k] == 10
              or match_skills[i][j][k] == 15):
          #if (match_skills[i][j][k] <= 10):
            words_coincidences.append(match_skills[i][j][k+1])
  return words_coincidences
```

```python
# This piece of code evaluates matches between job skills and candidate skills
# by using Levenshtein distance and only showing those whose values are between
# 0 and 10. 
# i.e: If job post skills are: 'Machine Learning' and 'Data Analytics', so we
# expetc to getting candidates with related skills as result

skills_filter = []
skills_len = []
for u in range(len(user_skills)):
  skills_filter.append(check_skills(job_skills=skills_job_post, candidate_skills=user_skills[u]))
  skills_len.append(len(skills_filter[u]))
  print(skills_filter[u],skills_len[u])
```

## Candidatos recomendados segun Skills requeridos

```python
df_skills = pd.DataFrame({'Skills': skills_filter, 'Skills_Matchs': skills_len})
df_skills['Index_Skills'] = df_skills.index
Skills_table = df_skills.sort_values('Skills_Matchs', ascending=False).head(20)
Skills_table
```

# Recomendación de candidatos por años de experiencia

```python
candidate_experiences = profiles['Experience']
candidate_experiences.tail(25)
```

```python
#  build a function that check candidate's experience by Levenshtein distance
def check_experience(job_experience, candidate_experiences):
  """ This function compares experience (years) of jobseekers 
  with job offer. Returns dataframe with Similarity level calculated by
  Levenshtein Distance.
  Example: 
  job_experience = `2 - 3 years`
  candidate_experiences = `['0-3 years', '2-5 years', '5-6 years']`
  """
  coincidences = []
  match_experiences = []
  for i in range(len(candidate_experiences)):
    match_experiences.append(dist.levenshtein(candidate_experiences[i], job_experience))
  df_data = pd.DataFrame(match_experiences, columns=['Similarity'])
  return df_data
```

```python
experience_filter = check_experience(job_experience=experience_job_post, 
                                     candidate_experiences=candidate_experiences)
# Display best matches by index and similarity
experience_filter['Index_Experience'] = experience_filter.index
Experience_table = experience_filter.sort_values(['Similarity', 'Index_Experience'], 
                                                 ascending=[True, True]).head(20)
Experience_table
```

```python
def check_title(job_title, candidate_titles):
  """ This function compares jobseekers title 
  with job offer's. Returns list with Similarity level calculated by
  Levenshtein Distance.
  Example: 
  job_title = `Technical Analyst - Data Science`
  candidate_titles = `['Business Analyst Data Science & Analytics', 
                        'Data Analyst - Data Science & Analytics', 
                        'Technical Analyst - Data Science']`
  """
  title_list = []
  coincidences = []
  match_titles = []
  data_temp = []
  for i in range(len(candidate_titles)):
    title_list.append(candidate_titles[i].split('-'))
  for j in range(len(title_list)):
    match_titles.append(dist.ilevenshtein(title_job_post,title_list[j]))
  for k in range(len(match_titles)):
    data_temp.append(match_titles[k])
  return data_temp
```

```python
# display results by titles
candidate_titles = profiles['Title']
title_filter = check_title(job_title=title_job_post, candidate_titles=candidate_titles)
df_title_filter = pd.DataFrame(title_filter, columns=['1st Title', '2nd Title', '3rd Title', '4th Title'])
df_title_filter.head(10)
```

```python
def get_distance_number(df):
  """ This function gets 1st title Levenshtein number from `check_title()`.
  Returns a data frame with all numbers
  """
  n = []
  for i in range(len(df)):
    n.append(df.iloc[i][0][0])
  df_out = pd.DataFrame(n)
  return df_out
```

```python
df_title_filter['Number'] = get_distance_number(df_title_filter)
```

```python
df_title_filter.head()
```

```python
# Add 'Competency' column and identify if candidates are suitable for job offer based on
# their titles 
df_title_filter['Competency'] = df_title_filter.Number.map(
    lambda x: 'Suitable' if x <= 20 else 'Non-suitable')
# we've selected all titles whose Levenshtein number is less than or equel to 20
```

```python
df_title_filter['Index_Title'] = df_title_filter.index
Title_table = df_title_filter.sort_values('Number', ascending=True).head(20)
Title_table
```

```python
# concatenate selected candidates (by indexes) and show them up as dataframe
df_table = pd.concat([Title_table['Index_Title'], 
                      Experience_table['Index_Experience'], 
                      Skills_table['Index_Skills']], 
                     axis=1)
df_table
```

```python
# search for coincidences and show results

# Create two list: `Index_Title` and `Index_Skills`
T_1 = list(Title_table['Index_Title'])
T_2 = list(Skills_table['Index_Skills'])

#  checks if `Index_Experience` is in `Index_Title` and saves it in variable
f_1 = df_table['Index_Experience'].isin(T_1)
Candidate_by_Experience = f_1.index[f_1]

#  checks if `Index_Skills` is in `Index_Title` and saves it in variable
f_2 = df_table['Index_Skills'].isin(T_1)
Candidate_by_Skills = f_2.index[f_2]

#  checks if `Index_Experience` is in `Index_Skills` and saves it in variable
f_3 = df_table['Index_Experience'].isin(T_2)
Candidate_by_skills_experience = f_3.index[f_3]
```
